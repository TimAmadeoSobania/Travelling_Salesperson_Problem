```python
import csv, pandas as pd, matplotlib.pyplot as plt, numpy as np
# Plot color definitions
burgundy = (.5647058823529412, .047058823529411764, .24705882352941178, 1)
green = (.3411764705882353, .7803921568627451, .5215686274509804, 1)
yellow = (1, .7647058823529411, 0, 1)
blue = (0, .7294117647058823, .6784313725490196)
orange = (1, .3411764705882353, .2, 1)
```

```python
col_names = []
data = []
fl = True

with open('results.csv', newline='') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in csvreader:
        if fl:
            col_names = row
            fl = False
        else:
            if len(row) == 7:
                data.append((int(row[0]),int(row[1]),float(row[2]),int(row[3])))#,str(row[4])))#,float(row[5]),str(row[6]))) 
        
df = pd.DataFrame(data = data, columns=col_names[:-3]).sort_values(by=[col_names[0]])

# Eigentlich .mean() - aber .median() filtert Ausreißer besser heraus (.min() für bestes Ergebnisse)
df_filter = df.groupby([col_names[1], col_names[3], col_names[0]]).median()
```

```python
# konkretes Beispiel: 12 Städte, 3 Prozesse, 4 Kerne
print(df.loc[df['num_procs'] == 3].loc[df['num_cities'] == 12].loc[df['num_cores'] == 4])
```

```python
cores = 64
cities_tested = df[col_names[0]].unique()
xy = np.arange(1,64)

for i in cities_tested:
    city_df = df_filter[np.in1d(df_filter.index.get_level_values(2),i)]
    city_df = city_df[np.in1d(city_df.index.get_level_values(1),cores)]
    
    t_ref = city_df[np.in1d(city_df.index.get_level_values(0),1)][col_names[2]].values[0]
    city_df["t_perfekt"] = t_ref / city_df.index.get_level_values(0)
    
    city_df['Speedup'] = t_ref / city_df[col_names[2]]
    
    
    city_df['Efficiency'] = city_df['Speedup'] / city_df.index.get_level_values(0)
    
    print("\n\n")
    print(city_df)
    print("\n\n")
    plt.figure(figsize=(10,5))
    plt.title("Number of cities tested: " + str(i) + " - Average duration")
    plt.ylabel("Time in s")
    plt.xlabel("Number of processes")
    plt.plot([0,65],[city_df[col_names[2]][0], city_df[col_names[2]][0]], c= (1, .3411764705882353, .2, .5))
    plt.scatter(list(city_df.index.get_level_values(0)), city_df[col_names[2]],c=[burgundy for i in city_df[col_names[2]]], label="Measured time")
    plt.scatter(list(city_df.index.get_level_values(0)), city_df["t_perfekt"],c=[orange for i in city_df[col_names[2]]], label="Perfect time")
    plt.legend(loc=1)
    plt.grid()
    plt.show()
    
    
    plt.figure(figsize=(10,5))
    plt.title("Number of cities tested: " + str(i) + " - Speedup")
    plt.ylabel("Speedup")
    plt.xlabel("Number of processes")
    plt.scatter(xy, xy,c=[blue for i in xy], label="Optimal speedup")
    plt.scatter(list(city_df.index.get_level_values(0)), city_df['Speedup'],c=[orange for i in city_df[col_names[2]]], label="Calculated speedup")
    plt.legend(loc=1)
    plt.grid()
    plt.show()
    
    plt.figure(figsize=(10,5))
    plt.title("Number of cities tested: " + str(i) + " - Efficiency")
    plt.ylabel("Efficiency")
    plt.xlabel("Number of processes")
    plt.scatter(list(city_df.index.get_level_values(0)), [1 for i in list(city_df.index.get_level_values(0))],c=[blue for i in list(city_df.index.get_level_values(0))], label="Maximal efficiency")
    plt.scatter(list(city_df.index.get_level_values(0)), city_df['Efficiency'],c=[orange for i in city_df[col_names[2]]], label="Efficiency")
    plt.legend(loc=5)
    plt.grid()
    plt.show()
```
